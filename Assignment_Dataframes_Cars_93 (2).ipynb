## Find out models of cars which have atleast 1 Airbag

```python
model_car_with_airbags = Cars[Cars["AirBags"] != 'NaN']["Model"]
model_car_with_airbags
```

## Find out all 5 seater cars that are of compact type

```python
Five_seater_car = Cars[(Cars.Passengers == 5) & (Cars.Type == "Compact")]
Five_seater_car
```

## Fetch only car type and Price and sort that data based on both Type wise average mileage on highway


```python
filtered_cols = ['Type', 'Price', 'MPG.highway']
average_mileage = Cars[filtered_cols].groupby('Type').mean()
sorted_data = average_mileage.sort_values('MPG.highway', ascending=False)
print(sorted_data)
```

## Find out MAke of all USA based Sporty cars

```python
USA_based_sporty_cars = Cars[(Cars.Origin == "USA") & (Cars.Type == "Sporty")][["Make"]]
USA_based_sporty_cars
```

## Find out the Manufacturer of the car having highest horsepower

```python
max_horsepower = Cars[Cars['Horsepower'] == Cars['Horsepower'].max()]
manufacturer = max_horsepower['Manufacturer'].values[0]
print(manufacturer)
```
