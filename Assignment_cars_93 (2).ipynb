##  Find out car having highest Mileage in city

```python
car_with_highest_city_mileage = Cars.loc[Cars['MPG.city'].max()]
car_with_highest_city_mileage
```

## Find out Models of cars which are having Price between 17 and 25

```python
car_model = Cars[(Cars.Price >17) & (Cars.Price < 25)]
car_model['Model']
```

##  Find out top 7 most fuel economic cars for highway

## Find out Models of cars having no AirBags and which are of compact type whose price is below 20.

```python
model = Cars[(Cars.AirBags.isnull()) & (Cars.Type == "Compact") & (Cars.Price < 20) ]
model['Model']
```

## Find out top 7 most fuel economic cars for city, show results in sorted manner based on Horsepower.

```python
most_fule_cars = Cars.sort_values(by=["Horsepower"],ascending = False)
```

```python
most_fule_cars[:7]
```

```python
city_fuel_sorted = Cars.nsmallest(7, 'MPG.city')[['Model', 'Horsepower', 'MPG.city']].sort_values(by='Horsepower')

print(city_fuel_sorted)
```

## Find out Cars having mileage of more than 15 miles in city

```python
high_city_milege = Cars[(Cars['MPG.city']>15)]
high_city_milege
```
